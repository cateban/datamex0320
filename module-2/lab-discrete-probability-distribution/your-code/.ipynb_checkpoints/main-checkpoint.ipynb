## Bernoulli's Distribution

In a basket containing 100 fruits, there are 60 apples and 40 oranges. 

A fruit is picked randomly from the basket. 

1. **What is the probability that the fruit is apple?** p
1. **What is the probability that it is an orange?** q

````python
from scipy.stats import binom
import matplotlib.pyplot as plt
from scipy.stats import poisson
import numpy as np
from math import factorial
````

````python
p=binom.pmf(1,1,0.6)
print('The probability of the randomly picked fruit to be an apple is {}'.format(p))
````

````python
q=binom.pmf(1,1,0.4)
print('The probability of the randomly picked fruit to be an orange is {}'.format(q))
````

Now we take a random sample of 20 fruits from the basket. After each fruit is taken, a new fruit of the same type is replaced in the basket. Therefore, every time we are taking 1 fruit from 100 fruits.

1. **What is the probability that the first 5 fruits are all apples and the next 15 fruits are all oranges?**

## Binomial Distribution

Similar to the above problem, this time again we randomly sample 20 fruits.

**What is the probability that this time the sample contains 5 apples and 15 oranges?**

Please note that this time the order of the fruits being picked does not matter.

````python
p=binom.pmf(5 , 20 , 0.6)
print('The probability of picking 5 apples is {}'.format(round(p,3)))
````

````python
x=[i for i in range(20)]  # numero de exitos
y=[binom.pmf(x, 20, 0.6) for x in x] # su probabilidad

plt.plot(x, y)
plt.title('Distribución Binomial',size=12,fontweight='bold')
plt.xlabel('Cantidad de manzanas', size=10)
plt.ylabel('Probabilidad', size =12);
````

In the lesson we also discussed CDF (cumulative distribution function). In the context of the above problem, **what is the probability that less than 5 fruits picked are apples?**

Write the formula and calculate the probability below.

````python
binom.cdf(5, 20, 0.6)
````

**Plot the PDF (probability distributiom function) for the above binomial distribution.**

*Hint: The range of x in a binomial distribution is from `0` to `n`.*

````python
x=[i for i in range(20)]  # numero de exitos
y=[binom.cdf(x, 20, 0.6) for x in x] # su probabilidad

plt.plot(x, y)
plt.title('Probabilidad acumulada',size=12,fontweight='bold')
plt.xlabel('Cantidad de manzanas', size=10)
plt.ylabel('Probabilidad', size =12);
````

## Poisson Distribution

In this challenge you are required to use the Math library in python to solve the problems.

In the UEFA champions league tournament, the average number of goals scored in a match is 2.3. 

**What is the probability that the number of goals scored in a match is 5?**

*Hint: Use the exp() in the math library*

And to calculate the factorial, you can use the following code:

```python
import math
math.factorial(n) # n is the number you want to find the factorial of
```

````python
def poisson_custom(x, lamb):  # pmf
    
    return np.exp(-lamb)*lamb**x/factorial(x)

uefa_c = poisson_custom(5,2.3)
uefa = poisson.pmf(5,2.3)

print(uefa_c)
print()
print(uefa)
````
