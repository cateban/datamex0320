## Obtaining the link

Wrote your codes in the cell below to obtain the data from the API endpoint you choose. A recap of the workflow:


```python
import json
import requests as req
import pandas as pd
from pandas.io.json import json_normalize
url = "https://swapi.co/api/people"
```

## Reading the data


```python
respuesta=req.get(url) #se establece la conexi√≥n
respuesta
```

```python
results=respuesta.json() #se pasa a formato json
```

```python
df_original=pd.DataFrame(results) #se pasa a data frame
df_original
```

```python
df = df_original.results.apply(pd.Series) #se extrae la informaci√≥n que me interesa

df
```

```python
df.height = df.height.astype(int)
df.mass = df.mass.astype(int)
```

```python
df.dtypes
```

## Bonus

Extract the following information from the data:

* The tallest

* The shortest

* The heaviest

* The lightest

## Tallest

```python
df[df.height == df.height.max()]
```

## Shortest

```python
df[df.height == df.height.min()]
```

## Heaviest

```python
df[df.mass == df.mass.max()]
```

## Lightest 

```python
df[df.mass == df.mass.min()]
```
